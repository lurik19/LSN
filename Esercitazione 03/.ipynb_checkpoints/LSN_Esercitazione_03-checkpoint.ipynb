## Esercizio 03.1

In questa esercitazione vogliamo mostrare che, simulando tramite metodi Monte Carlo opzioni Europee di tipo call e di tipo put, otteniamo una convergenza di $C\left[S(0),0\right]$ e $P\left[S(0),0\right]$ verso i valori calcolati analiticamente utilizzando la teoria di Black e Scholes. Questi valori analitici vengono calcolati nella cella successiva e verranno usati come confronto per i risultati numerici.

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
import scipy
from scipy.special import erf
from scipy import log, exp, sqrt

# data
S0 = 100.     # prezzo del titolo finanziario al tempo t = 0
K = 100.      # prezzo pattuito nel contratto
T = 1.        # data di chiusura del contratto
r = 0.1       # tasso di interesse (= drift dell'eq diff stocastica del GBM)
sigma = 0.25  # volatilità

t = 0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

 1) La prima richiesta dell'esercizio è di stimare $C\left[S(0),0\right]$ e $P\left[S(0),0\right]$ campionando direttamente il prezzo del titolo al tempo finale $S(t=T)$, dove $S(t)$ segue un moto browniano geometrico con drift $r$ e volatilità $\sigma$ ($GBM(r,\sigma^2)$). Per samplare $S(T)$ campioneremo quindi $W(T)$ da una PDF gaussiana con media $\mu=0$ e varianza $\sigma^2=T$ e inseriremo il risultato di questa estrazione nella seguente formula:
 
$$S(T) = S(0) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)T+\sigma W(T)\right]$$


Per ottenere il prezzo di call dell'opzione (Europea) a tempo $t=0$, applichiamo quindi un'evoluzione temporale al contrario (indietro nel tempo) al profitto che possiamo avere a $t=T$, cioè $\max(0, S(T)-K)$. Avremo quindi

$$C\left[S(0),0\right] = e^{-rT}\max(0, S(T)-K)$$

Ovviamente, trattandosi di una procedura Monte Carlo, faremo questa simulazione molte volte ($10^6$), utilizzando il metodo a blocchi (con $N=100$ blocchi) per ottenere il valore medio del prezzo di call a $t=0$ e la sua deviazione standard (della media).

Il grafico prodotto nella cella successiva mostra un buon accordo tra il valore simulato numericamente e il risultato teorico atteso (che abbiamo chiamato $(C_0)_{teo}$).

La procedura per calcolare il prezzo di put è molto simile, ma l'evoluzione temporale viene applicata questa volta a $\max(0, K - S(T))$, ottenendo in questo modo

$$P\left[S(0),0\right] = e^{-rT}\max(0, K - S(T))$$

Anche in questo caso la simulazione mostra un'ottima convergenza al valore teorico.

2) Anche nella seconda parte dell'esercitazione ci viene richiesto di simulare il valore di $C\left[S(0),0\right]$ e $P\left[S(0),0\right]$. Questa volta però dobbiamo ricavare $S(T)$ campionando l'intera traiettoria del moto browniano geometrico. In altri termini, dividiamo l'intervallo temporale $[0,T]$ in $100$ sotto intervalli e campioniamo il prezzo del titolo per ogni step temporale con la seguente formula:

$$S(t_{i+1}) = S(t_i) \exp\left[ \left(r-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    
dove $Z_{i+1}$ è una variabile aleatoria estratta da una PDF Gaussiana con $\mu=0$ e $\sigma=1$.

Una volta ottenuto $S(T)$ (cioè una volta simulati tutti i 100 passi) applichiamo le medesime formule della parte 1 per calcolare il prezzo di call e quello di put.

Anche questa volta abbiamo utilizzato il blocking method e ottenuto (per entrambi i prezzi) un buon accordo tra dati numerici e valori analitici.

```python
# Carico i dati da file
C0_prog, err_prog = ((np.genfromtxt('./Data/Es03.1/C0_es3.1.b.txt', usecols=i, autostrip=True)) for i in range(2))

x = np.arange(len(C0_prog))

# Disegno il plot
plt.ion()
plt.errorbar(x,C0_prog-call,yerr=err_prog)
plt.xlabel('#throws')
plt.ylabel(r'$<C_0> - (C_0)_{teo}$')
plt.grid(True)
plt.show()
```
