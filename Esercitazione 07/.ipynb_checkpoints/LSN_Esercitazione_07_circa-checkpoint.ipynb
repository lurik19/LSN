```python
M = 5 * 10**5
L = [10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000, 20000, 50000]

def error(AV,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return np.sqrt((AV2[n] - AV[n]**2)/n)

err = []

for k in range(len(L)):
    N = int(M/L[k])    
    ave = np.zeros(N)
    av2 = np.zeros(N)
    E_prog = np.zeros(N)
    E2_prog = np.zeros(N)
    errE_prog = np.zeros(N)

    for i in range(N):
        for j in range(L[k]):
            ave[i] += epot[j + i * L[k]]
        ave[i] /= L[k]
        av2[i] = (ave[i])**2

    for i in range(N):
        for j in range(i+1):
            E_prog[i] += ave[j]
            E2_prog[i] += av2[j]
        E_prog[i]/=(i+1)
        E2_prog[i]/=(i+1)
        errE_prog[i] = error(E_prog,E2_prog,i)
    
    err.append(errE_prog[N-1])
        
plt.plot(L, err)
plt.show()
```

```python
rNVT, gofrNVT, err_gofrNVT = ((np.genfromtxt('output.gave.0', usecols=i, autostrip=True)) for i in range(3))

rNVE, gofrNVE, err_gofrNVE = ((np.genfromtxt('./Esercitazione 04/output.gave.0', usecols=i, autostrip=True)) for i in range(3))

# Preparo la figura
fig, ax = plt.subplots(figsize=(10,6))
plt.xlabel('r')
plt.ylabel('g(r)')

# Zoom
x1, x2, y1, y2 = 1.9, 2.3, 1.0, 1.4
axins = zoomed_inset_axes(ax, zoom=5, bbox_to_anchor=(0, 0), loc=2)
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
plt.xticks(visible=False)
plt.yticks(visible=False)
mark_inset(ax, axins, loc1=2, loc2=1, fc="none", ec="0.5")

# Grafici
ax.errorbar(rNVT, gofrNVT, yerr=err_gofrNVT, label = 'NVT')
axins.errorbar(rNVT, gofrNVT, yerr=err_gofrNVT)

ax.errorbar(rNVE, gofrNVE, yerr=err_gofrNVE, label = 'NVE')
axins.errorbar(rNVE, gofrNVE, yerr=err_gofrNVE)

ax.legend(loc=1)
plt.show()
```
