## Esercizio 07.1

L'obiettivo di questo esercizio è capire quantitativamente quale sia la giusta taglia dei blocchi che utilizzeremo per il data blocking, affinchè i dati al loro interno non siano (eccessivamente) correlati.

Studieremo quindi l'autocorrelazione e l'andamento dell'errore in funzione di $L$ (taglia di un blocco) per le fasi solida, liquida e gassosa:

**N.B.**: in questo primo esercizio dovremo generare $5\times10^5$ dati e questo richiede molto tempo, per cui utilizzeremo il file <font face="Courier" size=3>main7.1.exe</font> che è una versione snellita del file che utilizzeremo negli esercizi successivi.

## Fase solida

Con questa fase il passo per l'algoritmo Metropolis, per avere un'accettazione media al $50\%$, è $\delta=0.12$. Il grafico dell'autocorrelazione mostra che per valori di $\tau>100$ si perda completamente la correlazione tra i dati; questo suggerisce che la taglia $L$ dei blocchi non possa essere più piccola di $100$. Il grafico seguente ci permette di avere una stima più precisa sul numero di dati per blocco In particolare si può osservare che l'errore satura dopo i $300$ dati per blocco: questo sarà quindi il valore che utilizzeremo per la fase solida nel corso di questa esercitazione.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset
from mpl_toolkits.axes_grid1.anchored_artists import AnchoredSizeBar
from IPython.display import clear_output

plt.rc('axes', labelsize=14)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=14)    # fontsize of the tick labels
plt.rc('legend', fontsize=14)    # fontsize of the legend
```

```python
import subprocess
def replace_line(file_name, line_num, text):
    with open(file_name, 'r') as file:
        data = file.readlines()
    data[line_num] = text
    with open(file_name, 'w') as file:
        file.writelines(data)

subprocess.call("cp input.solid input.dat", shell=True)
replace_line('input.dat', 5, '500000\n') # numero blocchi
replace_line('input.dat', 6, '1\n') # step per blocco
subprocess.call("make clean", shell=True)
subprocess.call("make main7.1.exe", shell=True)

#subprocess.call("./main7.1.exe")
#subprocess.call("mv -t Data/Es_7.1 output.epot.solid.0 output.pres.solid.0", shell=True)

print('Codice eseguito!')
```

```python
epot = (np.genfromtxt('./Data/Es_7.1/output.epot.solid.0', usecols=1, autostrip=True))
pres = (np.genfromtxt('./Data/Es_7.1/output.pres.solid.0', usecols=1, autostrip=True))

plt.subplots(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(epot)
plt.ylabel(r'$E_{pot}$')
plt.subplot(1,2,2)
plt.plot(pres)
plt.ylabel('Pressione')

plt.show()
```

```python
autocorrE = []
autocorrP = []

npoints = 200

for tau in range(npoints):
    clear_output(wait=True)
    print("Step ", tau + 1, "/", npoints)
    
    epot_1 = epot[:len(epot) - tau]
    epot_plus_tau = epot[tau:]
    
    pres_1 = pres[:len(pres) - tau]
    pres_plus_tau = pres[tau:]
    
    n = len(epot_1)
    autocorrE.append((sum(epot_1 * epot_plus_tau)/n - sum(epot_1) / n * sum(epot_plus_tau) / n)/np.std(epot)**2)
    autocorrP.append((sum(pres_1 * pres_plus_tau)/n - sum(pres_1) / n * sum(pres_plus_tau) / n)/np.std(pres)**2)
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(autocorrE)
plt.xlabel(r'$\tau$')
plt.ylabel(r'$AC_U(\tau)$')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(autocorrP)
plt.xlabel(r'$\tau$')
plt.ylabel(r'$AC_P(\tau)$')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Data blocking
err = []
M = len(epot)
listL = np.arange(500)*10 + 10
for L in listL:
    clear_output(wait=True)
    print("L = ", L, "/5000")
    N = int(M/L)
    ave = np.zeros(N)
    av2 = np.zeros(N)

    for i in range(N):
        for j in range(L):
            ave[i] += epot[j + i * L]
        ave[i] /= L
        av2[i] = (ave[i])**2
        
    E_prog = 0
    E2_prog = 0
    for j in range(N):
        E_prog += ave[j]
        E2_prog += av2[j]
    E_prog /= N
    E2_prog /= N
        
    err.append(np.sqrt((E2_prog - E_prog**2)/N))
```

```python
plt.subplots(figsize=(10,6))
plt.plot(listL, err)
plt.title('Errore stimato col data blocking', fontsize=16)
plt.xlabel('L')
plt.ylabel('errore')

plt.axvline(300, color='red')

plt.grid(True)
plt.show()
```

## Esercizio 07.2 

In questo esercizio ci viene richiesto di implementare nel codice il calcolo della funzione di distribuzione radiale (detta anche a coppie), $g(r)$ tra i valori del raggio $r \in [0,L_{box}/2]$, dove $L_{box}$ è il lato del box in cui stiamo compiendo la nostra simulazione.

In questa esercitazione stiamo studiando il caso di un materiale i cui atomi sono sottoposti a un potenziale che dipende dalla posizione di tutti gli atomi che compongono il materiale, $V\left(\vec r_1,...,\vec r_{npart}\right)$. Utilizzando questo si può calcolare la densità delle particelle che caratterizza il materiale citato.

$g\left(\vec r_1, \vec r_2\right)$ è definita come il rapporto tra la densità a due particelle del sistema studiato e la densità a due particelle del gas ideale, ovvero
$$g\left(\vec r_1, \vec r_2\right)=\frac{\rho^{(2)}\left(\vec r_1,\vec r_2\right)}{\rho^{(1)}\left(\vec r_1\right)\rho^{(1)}\left(\vec r_2\right)}$$

In particolare si può calcolare che vale la seguente relazione:
$$g(r)=g\left(|\vec r_1 - \vec r_2|\right) = \frac{1}{\rho n_{part}\Delta V(r)}\left\langle\sum_{j=1}^{n_{part}}\sum_{i\neq j=1}^{n_{part}}\delta\left[|\vec r| - |\vec r_i-\vec r_j\right]\right\rangle, \\ \Delta V(r)=\frac{4\pi}{3}\left[(r+dr)^3 -r^3\right]$$

che, a livello di codice, corrisponde a creare un istogramma e ad aggiungere due unità nel bin che corrisponde alla distanza $r$ ogni volta che due particelle sono a una distanza tra $r$ e $r+dr$. Dopo aver riempito l'istogramma, si normalizza dividendo ogni bin per $\rho n_{part}\Delta V(r)$. 

Dopo aver deciso di suddividere l'istogramma in 100 bin, il riempimento di questo è stato implementato nella funzione <font face="Fira Code" size=3>Measure()</font>. Il codice che ci è stato fornito misura già la distanza tra due particelle (necessaria per calcolare l'energia potenziale e la pressione); quindi noi abbiamo identificato il bin in cui cade questa misura della distanza e incrementato di due il valore dell'istogramma in questo bin.

Successivamente, nella funzione <font face="Fira Code" size=3>Averages()</font>, abbiamo normalizzato l'istogramma, calcolato (al variare di $r$) il valore medio di $g(r)$ e le sue incertezze (calcolati con il blocking method) e scritto i risultati nei file <font face="Courier" size=3>output.gofr.0</font> e <font face="Courier" size=3>output.gave.0</font>. Nel primo di questi file abbiamo scritto il numero del blocco considerato, il valore del raggio $r$, il valore medio di $g(r)$ e l'errore relativo. Invece, in <font face="Courier" size=3>output.gave.0</font> abbiamo scritto $r$, $\langle g(r)\rangle$ e $\sigma_g(r)$ per l'ultimo blocco (ovviamente per tutti i bin e quindi per i diversi valori di $r$).

Nell'esercizio 07.4 mostreremo diversi grafici di $g(r)$ e i valori li leggeremo da <font face="Courier" size=3>output.gave.0</font>.

```python
# Definiamo funzioni per convertire energia potenziale, pressione e raggio in unità SI
def SI_U(U_ist, U, errU):
    U_ist *= eps
    U *= eps
    errU *= eps

def SI_P(P_ist, P, errP): 
    P_ist *=eps/sigma**3
    P *=eps/sigma**3
    errP *=eps/sigma**3
    
def SI_R(r):
    r *= sigma

N, U_ist, U, errU = ((np.genfromtxt('./Data/output.epot.solid.0', usecols=i, autostrip=True)) for i in range(4))
N, P_ist, P, errP = ((np.genfromtxt('./Data/output.pres.solid.0', usecols=i, autostrip=True)) for i in range(4))

# Scriviamo le grandezze in unità SI
sigma = 0.34 * 10**-9 # m
k_b = 1.380649 * 10**-23 # J/K
eps = 120 * k_b # J
m = 39.948 * 1.66054*10**-27 # kg

SI_U (U_ist, U, errU)
SI_P (P_ist, P, errP)

# Grafici
plt.subplots(figsize=(16,5))
plt.subplot(1,2,1)
plt.errorbar(N, U, yerr=errU)
plt.xlabel('#throws')
plt.ylabel('U/N')
plt.grid(True)


plt.subplot(1,2,2)
plt.errorbar(N, P, yerr=errP)
plt.xlabel('#throws')
plt.ylabel('P')
plt.grid(True)
plt.show()
```

**Questo deve meravigliarci**, perché stiamo utilizzando due programmi molto diversi (uno che calcola approssimativamente la soluzione delle equazioni del moto, mentre l'altro che utilizza l'algoritmo di Metropolis per muovere le particelle che compongono il sistema), **ma è anche del tutto atteso e previsto dalla meccanica statistica**; infatti stiamo simulando lo stesso sistema: in un caso a temperatura fissata (NVT) e nell'altro caso a energia fissata (NVE). Se equilibriamo il sistema nell'ensemble microcanonico fino ad avere una temperatura che oscilla attorno alla temperatura target $T^*$ allora la fisica delle due simulazioni sarà la stessa e quindi otterremo la medesima $g(r)$.

```python
rNVT, gofrNVT, err_gofrNVT = ((np.genfromtxt('./Data/output.gave.solid.0', usecols=i, autostrip=True)) for i in range(3))
rNVE, gofrNVE, err_gofrNVE = ((np.genfromtxt('./Esercitazione_04/Data/output.gave.solid.0', usecols=i, autostrip=True)) for i in range(3))

SI_R(rNVE)
SI_R(rNVT)

# Preparo la figura
fig, ax = plt.subplots(figsize=(10,6))
plt.xlabel(r'$r$')
plt.ylabel(r'$g(r)$')

# Zoom
x1, x2, y1, y2 = 1.46*sigma, 1.58*sigma, 0.8, 1.0
axins = zoomed_inset_axes(ax, zoom=12, bbox_to_anchor=(0, 0), loc=2)
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
plt.xticks(visible=False)
plt.yticks(visible=False)
mark_inset(ax, axins, loc1=2, loc2=1, fc="none", ec="0.5")

# Grafici
ax.errorbar(rNVT, gofrNVT, yerr=err_gofrNVT, label = 'NVT')
axins.errorbar(rNVT, gofrNVT, yerr=err_gofrNVT)

ax.errorbar(rNVE, gofrNVE, yerr=err_gofrNVE, label = 'NVE')
axins.errorbar(rNVE, gofrNVE, yerr=err_gofrNVE)

ax.legend(loc=1)
ax.grid(True)
plt.show()
```

### Fase liquida
