## Esercizio 05.1

Lo scopo di questa esercitazione è implementare una prima applicazione dell'algoritmo di Metropolis, per campionare la distribuzione di probabilità della posizione dell'elettrone di un atomo di idrogeno. La PDF si ottiene prendendo il modulo quadro della funzione d'onda di interesse; in coordinate polari otteniamo:
$$p_{1,0,0}\left(r,\theta,\phi\right)=|\psi_{1,0,0}(r,\theta,\phi)|^2=\frac{a_0^{-3}}{\pi} e^{-2r/a_0}$$

In coordinate cartesiane quindi abbiamo, per l'**orbitale 1s**:

$$p_{1,0,0}\left(x,y,z\right)=\frac{a_0^{-3}}{\pi} \exp\left({-\frac{2}{a_0}\sqrt{x^2+y^2+z^2}}\right)$$

Per l'**orbitale 2pz** otteniamo la PDF:

$$p_{2,1,0}\left(x,y,z\right)=\frac{a_0^{-5}}{32\pi}z^2\exp\left({-\frac{1}{a_0}\sqrt{x^2+y^2+z^2}}\right)$$


**Ripetiamo che il nostro obiettivo è ora utilizzare l'algoritmo Metropolis per ottenere queste PDF**. In particolare inizieremo usando una <span style="color:blue">probabilità di transizione (di trial) $T\left(\vec x^*| \vec y\right)$</span> uniforme, ovvero che estragga uniformemente ogni componente di $\vec x^*$ tra $y_i - \delta$ e $y_i + \delta$, dove $y_i$ indica una componente del vettore $\vec y$ e $\delta$ è un parametro che dobbiamo specificare.

Dobbiamo innanzitutto chiederci da quale punto $(x_1,y_1,z_1)$ far partire la simulazione. Se partiamo da un punto lontano dall'origine, come ad esempio $(x_1=100,y_1=100,z_1=100)$, si può notare dal grafico sottostante che la PDF prodotta dall'algoritmo Metropolis non è quella tipica di un orbitale $1s$ (almeno per i primi step, finchè non viene raggiunto l'equilibrio).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
import subprocess

def replace_line(file_name, line_num, text):
    # Apro il file e leggo tutti i dati
    with open(file_name, 'r') as file:
        data = file.readlines()

    # Sostituisco la riga interessata
    data[line_num] = text

    # Riscrivo il file con le righe aggiornate
    with open(file_name, 'w') as file:
        file.writelines(data)

replace_line('input.dat', 0, '100\n')
subprocess.call("make main01.exe", shell=True)
subprocess.call("./main01.exe")

subprocess.call("mv XYZ_1s_100.out Data", shell = True)
subprocess.call("rm r_1s_100.txt r_2pz_100.txt XYZ_2pz_100.out", shell=True)

print("Codice eseguito!")
```

```python
M = 10**6
X = np.zeros(M)
Y = np.zeros(M)
Z = np.zeros(M)

fileXYZ = open('./Data/XYZ_1s_100.out', 'r')

for i in range(M):
    X[i], Y[i], Z[i] = fileXYZ.readline().split()

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

Questo comportamento ci suggerisce di scegliere il primo punto nello spazio 3D ($\vec x_1$) estraendolo da $T\left(\vec x_1| \vec 0\right)$. In questo modo $\vec x_1$ sarà vicino all'origine e quindi la PDF ottenuta sarà un orbitale $1s$ senza dover escludere nessun tiro che viene accettato dall'algoritmo di Metropolis. I successivi punti dell'esperimento verranno generati utilizzando $T\left(\vec x^*| \vec x_n\right)$.

Dobbiamo però introdurre anche la <span style="color:blue">probabilità di accettare la mossa, ovvero $A\left(\vec x^*| \vec x_n\right) = \min\left(1,p\left(\vec x^*\right)/p\left(\vec x_n\right)\right)$</span> che ha questa forma perchè $T$ è simmetrica, ovvero $T\left(\vec x^*| \vec x_n\right)=T\left(\vec x_n| \vec x^*\right)$. Un passo fondamentale dell'algoritmo Metropolis consiste nel generare (uniformemente) un numero casuale $r$ tra 0 e 1 per decidere se accettare la mossa o meno:
- se $r \leq A\left(\vec x^*| \vec x_n\right)$ allora accettiamo la mossa e $x_{n+1}=x^*$;
- se $r \geq A\left(\vec x^*| \vec y\right)$ la rifiutiamo e $x_{n+1} = x_n$.

Inizialmente utilizzeremo l'algoritmo di Metropolis per la PDF $p_{1,0,0}\left(x,y,z\right)$, per la quale il valore di $A$ è:

$$A\left(\vec x^*| \vec x_n\right) = \min\left(1,\frac{\frac{1}{\pi} \exp\left({-2\sqrt{|\vec x^*|^2}}\right)}{\frac{1}{\pi} \exp\left({-2\sqrt{|\vec x_n|^2}}\right)}\right)$$

dove abbiamo scritto le PDF in unità del raggio di Bohr.

Una parte integrante di questa esercitazione è consistita nella scelta del passo $\delta$ che caratterizza la probabilità di transizione. Infatti abbiamo dovuto scegliere questo valore in modo da rispettare la regola empirica secondo la quale debba essere accettato il 50% delle mosse proposte. In questo caso, dopo aver avviato qualche volta la simulazione ci siamo resi conto che un valore ideale è $\delta=1.23$, per il quale $\langle A\rangle\simeq49.9\%$

Ripetendo la procedura $M=10^6$ volte otteniamo una vettore di $M$ posizioni nello spazio tridimensionale. A questo insieme di vettori applicheremo il data bocking (dividendo gli $M$ tiri in $N=100$ blocchi) per ottenere il valore medio del raggio, $\langle r\rangle$ e la sua incertezza. Ci si può chiedere quali siano le ragioni che ci hanno portato a scegliere $N=100$. Il nostro codice (presente nel file <font face="Courier" size=3>main01.cpp</font>) calcola il valore medio dell'accettazione $\langle A\rangle$ per ogni blocco e stampa i valori massimo e minimo che questa grandezza assume tra gli $N$ blocchi. L'idea è quella di dividere gli $M$ tiri in $N$ blocchi, ciascuno dei quali deve essere caratterizzato da $\langle A\rangle \simeq50\%$, in modo che i lanci all'interno di ogni blocco non siano troppo correlati. (**CREDO!!!**)
Scegliere un numero di blocchi troppo grande, come per esempio $N=10^4$ restituisce $\min\langle A\rangle \simeq 32\%$ e $\max\langle A\rangle \simeq 64\%$, che sono valori non desiderabili. Scegliendo invece $N=100$ otteniamo $49\% \leq \langle A\rangle \leq 51\%$. (**<font color="red">PENSACI</font>**)


Nel primo plot presente nelle celle successive si può osservare che, utilizzando l'algoritmo Metropolis, abbiamo ottenuto una distribuzione di probabilità tipica degli orbitali 1s (che è quella che stavamo cercando). Inoltre, nel grafico sottostante, si osserva che il valore medio di $r$ converge al valore teorico atteso, ovvero $3/2$ (in unità di $a_0$).

```python
replace_line('input.dat', 0, '0\n')
subprocess.call("./main01.exe")

subprocess.call("mv -t Data XYZ_1s_0.out r_1s_0.txt XYZ_2pz_0.out r_2pz_0.txt", shell = True)


print("Codice eseguito!")
```

```python
M = 10**6
X = np.zeros(M)
Y = np.zeros(M)
Z = np.zeros(M)

fileXYZ = open('./Data/XYZ_1s_0.out', 'r')

for i in range(M):
    X[i], Y[i], Z[i] = fileXYZ.readline().split()

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
M = 10**6               # Numero di  tiri
N = 100                 # Numero di blocchi
L = int(M/N)            # Numbero di esperimenti (o tiri) per blocco

x = (np.arange(N))*L      # [0,L,2*L,...,(N-1)*L]

r_prog = np.zeros(N)
err_prog = np.zeros(N)

# Carico i dati da file
file1 = open('./Data/r_1s_0.txt', 'r')

for i in range(N):
    r_prog[i], err_prog[i] = file1.readline().split()

# Disegno il plot
plt.errorbar(x,r_prog-1.5,yerr=err_prog)
plt.xlabel('#throws')
plt.ylabel(r'$<r>-3/2$')
plt.grid(True)
plt.show()
```

Ora studiamo il caso dell' orbitale 2pz. La probabilità di accettazione è in questo caso:

$$A\left(\vec x^*| \vec x_n\right) = \min\left(1,\frac{\frac{1}{32\pi}(z^*)^2\exp\left({-\sqrt{|\vec x^*|^2}}\right)}{\frac{1}{32\pi}(z_n)^2\exp\left({-\sqrt{|\vec x_n|^2}}\right)}\right)$$

In questo caso i punti ottenuti usando l'algoritmo seguono la distribuzione spaziale di un orbitale 2pz e la simulazione converge verso il valore teorico, $\langle r\rangle = 5 a_0$.

Per questa PDF la scelta ottimale del passo è stata $\delta=2.98$, che corrisponde a $\langle A\rangle \simeq 50\%$. Anche studiando questo orbitale la scelta $N=100$ restituisce dei valori di $\langle A\rangle$ accumulati attorno al $50\%$ per tutti i blocchi.

```python
fileXYZ = open('./Data/XYZ_2pz_0.out', 'r')
M = 10**6
X = np.zeros(M)
Y = np.zeros(M)
Z = np.zeros(M)

for i in range(M):
    X[i], Y[i], Z[i] = fileXYZ.readline().split()

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

```python
M = 10**6               # Numero di  tiri
N = 100                 # Numero di blocchi
L = int(M/N)            # Numbero di esperimenti (o tiri) per blocco

x = np.arange(N)*L      # [0,L,2*L,...,(N-1)*L]

r_prog = np.zeros(N)
err_prog = np.zeros(N)

# Carico i dati da file
file1 = open('./Data/r_2pz_0.txt', 'r')

for i in range(N):
    r_prog[i], err_prog[i] = file1.readline().split()

# Disegno il plot
plt.errorbar(x,r_prog-5,yerr=err_prog)
plt.xlabel('#throws')
plt.ylabel(r'$<r>-5$')
plt.grid(True)
plt.show()
```

I grafici mostrati successivamente mostrano che la situazione rimane qualitativamente la stessa se utilizziamo come probabilità di transizione $T$ una PDF Gaussiana per ogni coordinata $x,y,z$, ovvero

$$T\left((x^*)_i|(x_n)_i\right)=\mathcal{N}\left( (x_n)_i,\delta\right)$$

La convergenza ai valori teorici di $\langle r \rangle$ è rispettata anche con questa scelta. **DÌ QUALCOSA DI PIÙ**.

Le scelte per $\delta$ in questo caso sono state:
- Orbitale $1s$: $\delta = 0.76 => \langle A\rangle \simeq50.0\%$;
- Orbitale $2pz$: $\delta = 1.87 => \langle A\rangle \simeq50.1\%$

Anche scegliendo $T$ gaussiana, $N=100$ è una scelta ottimale per il numero di blocchi.
