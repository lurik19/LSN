## Esercizio 10.1

In questo esercizio ci viene richiesto di implementare la tecnica del **simulated annealing** nel nostro algoritmo genetico. In particolare non utilizzeremo più un'intera popolazione per stimare il percorso compiuto dal TS, bensì useremo un algoritmo di Metropolis.

Quello che faremo è utilizzare una serie di temperature (o equivalentemente una serie di valori di $\beta$) e compiere un certo numero di mosse di Metropolis per ogni temperatura. Partiamo da una temperatura sufficientemente alta e, una volta compiuto il numero di mosse desiderato ($n_\text{moves}$), la abbasseremo e ripeteremo il procedimento. Questo algoritmo dovrebbe portarci a convergere al valore desiderato dell'energia (la funzione di costo).

Nella simulazione che presenteremo abbiamo deciso di considerare $\beta \in [1, 1000]$ e $n_\text{moves}=50$. È affascinante notare come, con così poche mosse di accettazione per ogni temperatura, si possa ottenere un'ottima convergenza!

Iniziamo discutendo il caso delle **città distribuite casualmente su una circonferenza**. Dai grafici presentati qui sotto possiamo osservare che, anche con il simulated annealing, come per l'esercitazione precedente, otteniamo il più breve tragitto che il commesso viaggiatore può percorrere (ovvero lungo la circonferenza). Studiando il plot successivo (ovvero l'andamento della lunghezza del migliore path in funzione di $\beta$) notiamo inoltre che, al crescere di $\beta$ (ovvero al decrescere della temperatura) la lunghezza del tragitto del TS decresce effettivamente, andando a convergere al valore minimo, che è uguale al valore di $L_1$ ottenuto all'ultima generazione dell'algoritmo genetico (v. <a href="../Esercitazione 09/LSN_Esercitazione_09.ipynb">Esercitazione 9</a>), anche questo presente nel grafico.

Confrontiamo ora l'efficienza di questi due algoritmi:
- con l'algoritmo genetico $L_1$ satura dopo circa 100 generazione, che significa valutare <span style="color:blue">$100*n_\text{pop}=10^5$ individui</span>;
- con il simulated annealing abbiamo convergenza per $\beta \simeq 600$, ovvero dopo aver valutato <span style="color:blue">$600*n_\text{moves}=3*10^4$ individui</span>.

Questo ci suggerisce che il simulated annealing sia più efficiente rispetto all'algoritmo genetico.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rc('axes', labelsize=14)     # fontsize of the x and y labels
plt.rc('xtick', labelsize=14)    # fontsize of the tick labels
plt.rc('ytick', labelsize=14)    # fontsize of the tick labels
plt.rc('legend', fontsize=14)    # fontsize of the legend
```

```python
import subprocess
def replace_line(file_name, line_num, text):
    with open(file_name, 'r') as file:
        data = file.readlines()
    data[line_num] = text
    with open(file_name, 'w') as file:
        file.writelines(data)
        
replace_line('input.dat', 0, '0\n') # città sulla circonferenza
subprocess.call("make clean", shell=True)
subprocess.call("make", shell=True)
subprocess.call("./main.exe")

print('Codice eseguito!')
```

```python
# Grafico delle posizioni delle città
x,y = (np.genfromtxt('./Data/cities_circ.dat', usecols=i, autostrip=True) for i in range(2))

plt.subplots(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(x,y, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Città da visitare', fontsize=16)

# Grafico del percorso seguito dal TS
x,y = (np.genfromtxt('./Data/cities_final_circ.dat', usecols=i, autostrip=True) for i in range(2))

plt.subplot(1,2,2)
plt.plot(x,y, 'o-', label='Tragitto')
plt.plot(x[0],y[0],'*', markersize=15, label='Casa')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Tragitto del commesso viaggiatore', fontsize=16)
plt.legend(loc=0)

plt.tight_layout()
plt.show()
```

```python
beta, energy = (np.genfromtxt('./Data/Length_circ.dat', usecols=i, autostrip=True) for i in range(2))

plt.subplots(figsize=(10,6))
plt.plot(beta, energy, label='Migliore lunghezza SA')

fitness = np.genfromtxt('../Esercitazione 09/Data/cost_circ.dat', usecols=1, autostrip=True)

plt.axhline(fitness[-1], color='C2', label="Migliore lunghezza GA")
plt.axhline(2*np.pi, label=r'Lunghezza attesa $=2\pi$', color='C1')

plt.xlabel(r'$\beta$')
plt.legend(loc=0)
plt.grid(True)
plt.show()
```

## Esercizio 10.2 

In questo esercizio recuperiamo l'algoritmo genetico (quindi il crossover e il fatto di avere una popolazione) e parallelizziamo questa procedura.

Questo, in altri termini, significa utilizzare le librerie **MPI**. Tramite esse possiamo, con pochi comandi, parallelizzare un programma (con il modello SPMD) ed eseguirlo su più core del nostro PC (4 nel nostro caso).

Bisogna però implementare una novità, ovvero il fatto che, dopo $n_\text{migr}$ generazioni, i 4 processi (o ***continenti***) si scambieranno il loro migliore individuo in maniera casuale. La domanda che ci poniamo è se la parallelizzazione ci consente di ottenere risultati migliori.

Alcune accortezze a livello di codice:
- per avere le stesse città in tutti e 4 i processi bisogna utilizzare la funzione <font face="Fira Code" size=3>MPI_Bcast</font> per copiare le città da un processo *root* a tutti gli altri processi;
- per ogni processo bisogna inizializzare le librerie di numeri random con dei semi diversi, per avere risultati diversi.

Commentiamo ora i **risultati**: dopo 200 generazioni, utilizzando $n_\text{migr}=10$ osserviamo che i 4 processi arrivano esattamente alla medesima soluzione, che è anche la stessa ottenuta nell' Esercitazione 9.

Anche per quanto riguarda l'andamento di $L_1$ al crescere delle generazioni osserviamo che la situazione è quantitativamente molto simile a quella osservata per un singolo processo, nell'esercitazione precedente. Questo fatto non è soprendente, infatti una migrazione può essere vista come una mutazione che avviene nel passaggio da una generazione a quella successiva, lasciando quindi (statisticamente) invariata la situazione. Un'osservazione più attenta dei grafici ci fa notare che la soluzione ottima ($L_1$ minimo) viene ottenuta con un numero di generazioni leggermente più basso nel caso in cui consideriamo le migrazioni.
